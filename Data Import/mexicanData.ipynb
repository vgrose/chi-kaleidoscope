```python
master_data = data1+data2+data3+data4+data5+data6+data7+data8+data9+data10+data11+data12+data13+data14+data15+data16+data17+data18+data19
# master_data
mexican_df = pd.DataFrame({'Name': master_data})
# pizza_df.head(10)
mexican2_df = mexican_df.drop(mexican_df.index[0])

mexican2_df.head(10)
```

```python
##################################################################################
############################### Clean the data ###################################
```

```python
##Extract the ZipCode and Telephone Number info from the mexican2_df["Name"] column 
new = mexican2_df["Name"].str.split("IL",n=1,expand = True)

# making separate "ziptel" column from new data frame 
mexican2_df["name2"]= new[0] 
mexican2_df["ziptel"]=new[1]

mexican2_df.drop(columns = ["Name"], inplace = True)

mexican2_df['Phone Number'] = mexican2_df['ziptel'].str[-12:]
# print(pizza2_df['Phone Number'])
mexican2_df['Zip Code'] = mexican2_df['ziptel'].str[:6]
# print(pizza2_df['Zip Code'])
mexican2_df.drop(columns =["ziptel"], inplace = True)
```

```python
mexican2_df.head(10)
```

```python
##Extract the City and Telephone Number info from the mexican2_df["Name"] column 
new = mexican2_df["name2"].str.split("Chicago",n=1,expand = True)

# making separate "name3" column from new data frame 
mexican2_df["name3"]= new[0] 
mexican2_df["chicago"]= new[1]

mexican2_df.drop(columns = ["name2"], inplace = True)
mexican2_df.drop(columns = ["chicago"], inplace = True)

#Choose the most important columns for the next part of the ETL Project
mexican2_df = mexican2_df[["name3","Phone Number","Zip Code"]]
mexican2_df.head(10)
```

```python
mexican2_df.to_excel('../../Desktop/export_dataframe.xlsx', index = None, header=True) #Don't forget to add '.xlsx' at the end of the path
```

```python
mexican3_df = pd.read_csv("../../Desktop/mexicanData.csv") 
```

```python
mexican3_df.head()
```

```python
##Extract and separate "Name" and "Address" from the mexican3_df["name3"] column

# dropping null value columns to avoid errors 
mexican3_df.dropna(inplace = True) 
  
# new dataframe with split value columns 
new = mexican3_df["name3"].str.split(",", n = 1, expand = True) 
  
# making separate "Name" column from new data frame 
mexican3_df["Name"]= new[0] 
  
# making separate "Address" column from new data frame 
mexican3_df["Address"]= new[1] 

# Dropping old "name3" column
mexican3_df.drop(columns =["name3"], inplace = True) 

mexican3_df.head()
```

```python
#Choose the most important columns for the next part of the ETL Project
mexicanFinal_df = mexican3_df[["Name","Address","Phone Number","Zip Code"]]
#Rename column names so that they are SQL feiendly
mexicanFinal_df.columns=["Name","Address","Phone_Number","Zip_Code"]
mexicanFinal_df.head()
mexicanFinal_df.reset_index(drop = True)
```
