```python
import pandas as pd
```

```python
url="https://chicago.curbed.com/maps/map-haunted-places"
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
c-mapstack_info
```

```python
results = soup.find_all(class_="c-mapstack__card ")
results
```

```python
gkey = "AIzaSyDlqMW5JotVMO1brg3HsPp7x51a7__cfTk"
base_url = "https://maps.googleapis.com/maps/api/geocode/json"
haunted=[]

for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        name = result.find(class_ = "c-mapstack__card-index").text
        address = result.find(class_='c-mapstack__address').text
        # Identify and return price of listing
        blurb = result.find("p").text

        # Print results only if title, price, and link are available
        if (name and address and blurb):
            print('-------------')
            print(name)
            new_dict={}
            new_dict["Name"]=name
            print(address)
            new_dict["Address"]=address
            print(blurb)
            new_dict["Blurb"]=blurb
        
            params = {
                "address": address,
                "key": gkey
            }
            geo = requests.get(base_url, params).json()
            lat = geo["results"][0]["geometry"]["location"]["lat"]
            print(lat)
            lng = geo["results"][0]["geometry"]["location"]["lng"]
            print(lng)
            new_dict["Latitude"] = lat
            new_dict["Longitude"] = lng
            print(new_dict)
            haunted.append(new_dict)
            
    except AttributeError as e:
        print()
        
haunted_df=pd.DataFrame(haunted)
haunted_df.head()
```

```python
haunted_df.to_csv("haunted.csv")
```
